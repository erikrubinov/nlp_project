```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv('../reports/classification_report_1.csv')
```

```python
data 
```

```python
# rename first column
data.rename(columns={'Unnamed: 0': 'type'}, inplace=True)
```

```python
data
```

```python
# remove all where type not "weighted avg"
data = data[data['type'] == 'weighted avg']
data
```

```python
sns.barplot(x='classifier', y='f1-score', hue="preprocess_function", data=data)
```

```python
sns.barplot(x='vectorizer', y='f1-score', hue="preprocess_function", data=data)
```

```python
# merge ngram_lower_bound and ngram_upper_bound into one column
data['ngram_lower_bound'] = data['ngram_lower_bound'].astype(str)
data['ngram_upper_bound'] = data['ngram_upper_bound'].astype(str)
data['ngram'] = data['ngram_lower_bound'] + '-' + data['ngram_upper_bound']
sns.barplot(x='ngram', y='f1-score', hue="preprocess_function", data=data)
```
